```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

$\hat{\phi}(k)/\hat{\phi}_0  = \exp(-(\sigma k_0)^2((k/k_0)^2-1))$

```python
var = 0.01
k = np.arange(1-6*np.sqrt(var),1+6*np.sqrt(var),12*np.sqrt(var)/1000)
phi_h = np.exp(-(k-1)**2/var)
```

```python
fig = plt.figure()
plt.plot(k,phi_h)
plt.xlabel(r'$\frac{k}{k_0}$',size=18)
plt.ylabel(r'$\frac{\hat{\phi}}{\hat{\phi}_0}$',rotation=0,size=18)
plt.xlim(-6,8)
plt.ylim(0,1)
plt.savefig('11.png')
plt.show()
```

```python
x = np.arange(-24/np.sqrt(var),24/np.sqrt(var),48/(np.sqrt(var)*10000))
phi = np.exp(-(var*x)**2/4)*np.cos(x)
fig = plt.figure()
plt.plot(x,phi)
plt.xlabel(r'$k_0 x$',size=18)
plt.ylabel(r'$\frac{\phi}{\phi_0}$',rotation=0,size=18)
plt.savefig('12.png')
plt.show()
```
